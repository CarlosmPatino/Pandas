## como manejar datos nulos

```python
dicc = {'col1': [1,2,3,np.nan],
 'col2':[4,np.nan,6,7],
 'col3':['a','b','c', None]}
```

```python
df = pd.DataFrame(dicc)
df
```

```python
#como saber donde hay valores nulos, con el siguiente comando nos devuelve una matriz de true and false de donde se encuentran
df.isnull()
```

```python
df.isnull()*1 #resultado numerico
```

```python
#llena las casillas nan con lo que coloquemos en este caso missing
df.fillna('Missing')
```

```python
#podemos poner la media
df.fillna(df.mean()) # pero como hay valores str no me deja 
df.interpolate() # rellena el valor faltante asumiendo la serie  
```

```python
df.dropna() #elimina todos las filas con valores nulos
```

## filtrado por condiciones


```python
df_Books = pd.read_csv('best_seller.csv', sep = ',', header = 0)
df_Books.head()
```

```python
# obtenemos una lista booleana de quienes cumplen esta condicion 
mayor_A_2016 = df_Books ['Year'] > 2016
```

```python
#pasamos esta lista al data frame y obtenemos la informacion de los que la cumplen
df_Books[mayor_A_2016]
```

```python
#si queremos negar el resultado, que me muestre los ocntrario hacemos los siguiente
df_Books[~mayor_A_2016]
```

```python
#tambien podemos filtrar de la siguiente forma 
df_Books[df_Books['Author'].str.contains('Palacio')] #str contains me indica cuando una lista contiene ese valor y neme devuelve un true
```

## concatenar

```python
df1 = pd.DataFrame({'A': ['A0','A1','A2','A3'],
                    'B': ['B0','B1','B2','B3'],
                    'C': ['C0','C1','C2','C3'],
                    'D': ['D0','D1','D2','D3']})

df2 = pd.DataFrame({'A': ['A4','A5','A6','A7'],
                    'B': ['B4','B5','B6','B7'],
                    'C': ['C4','C5','C6','C7'],
                    'D': ['D4','D5','D6','D7']})
```

```python
#concatenar dos data frame ignorando el indice
pd.concat([df1,df2], ignore_index=True)
```

```python
#concatenar por columnas
pd.concat([df1,df2], axis= 1)
```

## MERGE

```python
izq = pd.DataFrame({'key': ['K0','K1','K2','K3'],
                    'A': ['A0','A1','A2','A3'],
                    'B': ['B0','B1','B2','B3']})


der = pd.DataFrame({'key': ['K0','K1','K2','K3'],
                    'C': ['C0','C1','C2','C3'],
                    'D': ['D0','D1','D2','D3']})
```

```python
#inner join, especificamos que columna queremos que se haga el match
izq.merge(der, on = 'key')
```

```python
izq = pd.DataFrame({'key': ['K0','K1','K2','K3'],
                    'A': ['A0','A1','A2','A3'],
                    'B': ['B0','B1','B2','B3']})


der = pd.DataFrame({'key_1': ['K0','K1','K2','K3'],
                    'C': ['C0','C1','C2','C3'],
                    'D': ['D0','D1','D2','D3']})
```

```python
#si no concuerda el nombre de la columna hacemos los siguiente
izq.merge(der, left_on = 'key', right_on='key_1')
```

```python
izq = pd.DataFrame({'key': ['K0','K1','K2','K3'],
                    'A': ['A0','A1','A2','A3'],
                    'B': ['B0','B1','B2','B3']})


der = pd.DataFrame({'key_1': ['K0','K1','K2',np.nan],
                    'C': ['C0','C1','C2','C3'],
                    'D': ['D0','D1','D2','D3']})
```

```python
#para especificar que tipo de union queremos utilizamos how
izq.merge(der, left_on = 'key', right_on='key_1', how='left')
```

```python
izq.merge(der, left_on = 'key', right_on='key_1', how='right')
```

```python
izq.merge(der, left_on = 'key', right_on='key_1', how='inner')
```

## JOIN INDEX MATCH

```python
izq = pd.DataFrame({'A': ['A0','A1','A2','A3'],
                    'B': ['B0','B1','B2','B3']},
                    index=['k0','k1','k2','k4'])


der = pd.DataFrame({'C': ['C0','C1','C2','C3'],
                    'D': ['D0','D1','D2','D3']},
                    index=['k0','k2','k3','k4'])
```

```python
#tiene sintaxix parecida a merge
izq.join(der, how='inner')
```

```python
#me trae todo
izq.join(der, how='outer')
```

```python
izq.join(der, how='left')
```
